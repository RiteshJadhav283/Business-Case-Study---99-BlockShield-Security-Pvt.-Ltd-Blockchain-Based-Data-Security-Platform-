```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('blockshield_monitoring.csv')
print('Rows loaded:', len(df))
```

### Mean Values

```python
print('Variable 2 — Average System Uptime        :', round(df['uptime_pct'].mean(), 4), '%')
print('Variable 3 — Avg Unauthorized Attempts/Day:', round(df['unauthorized_attempts'].mean(), 1))
print('Variable 6 — Avg Incident Response Time   :', round(df['incident_response_min'].mean(), 2), 'minutes')
```

### Graph — System Uptime over 365 Days

```python
plt.figure(figsize=(10, 4))
plt.plot(df['uptime_pct'], color='steelblue', linewidth=0.9)
plt.axhline(df['uptime_pct'].mean(), color='red', linestyle='--',
            label=f'Mean: {df["uptime_pct"].mean():.4f}%')
plt.axhline(99.99, color='orange', linestyle='--', label='Target: 99.99%')
plt.title('System Uptime — 365 Days')
plt.xlabel('Day')
plt.ylabel('Uptime %')
plt.ylim(98.5, 100.1)
plt.legend()
plt.tight_layout()
plt.show()
```

### Graph — Unauthorized Attempts per Day

```python
plt.figure(figsize=(8, 4))
plt.hist(df['unauthorized_attempts'], bins=20, color='orange', edgecolor='white')
plt.axvline(df['unauthorized_attempts'].mean(), color='red', linestyle='--',
            label=f'Mean: {df["unauthorized_attempts"].mean():.1f}/day')
plt.axvline(50, color='darkred', linestyle='--', label='Alert threshold: 50/hr')
plt.title('Unauthorized Access Attempts per Day')
plt.xlabel('Attempts')
plt.ylabel('Number of Days')
plt.legend()
plt.tight_layout()
plt.show()
```

### Graph — Incident Response Time

```python
plt.figure(figsize=(8, 4))
plt.hist(df['incident_response_min'], bins=20, color='purple', edgecolor='white', alpha=0.8)
plt.axvline(df['incident_response_min'].mean(), color='red', linestyle='--',
            label=f'Mean: {df["incident_response_min"].mean():.2f} min')
plt.axvline(5, color='orange', linestyle='--', label='Target: < 5 min')
plt.title('Incident Response Time Distribution')
plt.xlabel('Minutes')
plt.ylabel('Number of Days')
plt.legend()
plt.tight_layout()
plt.show()
```

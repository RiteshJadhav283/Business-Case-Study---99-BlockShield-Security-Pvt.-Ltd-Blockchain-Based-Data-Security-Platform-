```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('blockshield_transactions.csv')
print('Rows loaded:', len(df))
```

### Mean Values

```python
print('Variable 1 — Encryption Success Rate :', round(df['encryption_success'].mean() * 100, 4), '%')
print('Variable 4 — Average TPS             :', round(df['tps'].mean(), 2), 'transactions/sec')
print('Variable 5 — Tamper Detected Count   :', df['tamper_detected'].sum(), 'out of', len(df))
```

### Graph — TPS Distribution

```python
plt.figure(figsize=(8, 4))
plt.hist(df['tps'], bins=30, color='steelblue', edgecolor='white')
plt.axvline(df['tps'].mean(), color='red', linestyle='--', label=f'Mean: {df["tps"].mean():.1f} TPS')
plt.axvline(1000, color='orange', linestyle='--', label='Target: 1000 TPS')
plt.title('Transaction Processing Speed (TPS)')
plt.xlabel('TPS')
plt.ylabel('Count')
plt.legend()
plt.tight_layout()
plt.show()
```

### Graph — Encryption Success vs Failed

```python
counts = df['encryption_success'].value_counts()
plt.figure(figsize=(5, 4))
plt.bar(['Success', 'Failed'], [counts.get(True, 0), counts.get(False, 0)],
        color=['green', 'red'], edgecolor='white')
plt.title('Encryption Success vs Failed')
plt.ylabel('Number of Transactions')
for i, v in enumerate([counts.get(True, 0), counts.get(False, 0)]):
    plt.text(i, v + 2, str(v), ha='center', fontweight='bold')
plt.tight_layout()
plt.show()
```
